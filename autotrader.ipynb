```python
from splinter import Browser
from bs4 import BeautifulSoup 
import json
```

```python
browser = Browser('chrome')
```

```python
url = 'https://www.autotrader.com/cars-for-sale/all-cars?zip=78741&makeCodeList=FORD&modelCodeList=MUST'
browser.visit(url)
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.get_text())
```

```python
mustangs = soup.find_all('div', class_='item-card-body margin-bottom-auto')
mustangs
```

```python
mustangList = []
```

```python
for mustang in mustangs:

    title = mustang.find('div', class_= "display-flex justify-content-between").text.strip()
    price = mustang.find('span', class_='first-price').text.strip()

    if (title and price):
        print('------------------')
        print(title)
        print(price)
        
    mustangDict = {
        'title':title,
        'preview':price
    }
    
    mustangList.append(mustangDict)
```

```python
browser.links.find_by_partial_text("Next Page").click()
```
