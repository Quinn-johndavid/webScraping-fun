```python
def get_Scrapes():
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    usedcars = soup.find('div', id = 'lvrp-results-wrapper')
    
  
    
    # Find all the articles in the main area of the web page
    usedcarDetails = usedcars.find_all("div", class_= 'hit-content')


    for detail in usedcarDetails:

        title = detail.find('span', class_= "title-bottom").text.strip()
        price = detail.find('div', class_='hit-price__value').text.strip()

        if (title and price):
            print('------------------')
            print(title)
            print(price)
        
        usedDict = {
            'title':title,
            'preview':price
        }
    
        usedcarsList.append(usedDict)
get_Scrapes()
```

```python
browser.links.find_by_partial_text("Next").click()
```

```python
for _ in range(5):
    get_Scrapes()
    browser.links.find_by_partial_text("Next").click()
```
